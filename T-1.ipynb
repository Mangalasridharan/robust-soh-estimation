```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
time = df['time/s']
applied_current = df['<I>/mA']
print(time.shape)
print(applied_current.shape)
```

```python
plt.figure(figsize=(10,8))
plt.plot(time,applied_current)  
plt.xlabel("Time(s)")
plt.ylabel("Appliedcurrent(I/mA)")
plt.grid()
plt.legend(['Applied Current'])
plt.title("Applied current vs Time")
```
